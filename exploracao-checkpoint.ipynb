```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../data/raw/SMSSpamCollection", sep="\t", header=None, names=["label", "text"])
data.head()
```

```python
print("Total de mensagens:", len(data))
print(data["label"].value_counts())

sns.countplot(x="label", data=data)
plt.title("Distribuição de mensagens (Spam vs Ham)")
plt.show()
```

```python
data["length"] = data["text"].apply(len)

print("Tamanho médio por classe:")
print(data.groupby("label")["length"].mean())

sns.histplot(data=data, x="length", hue="label", bins=50, kde=True)
plt.title("Distribuição do tamanho das mensagens")
plt.show()
```

```python
print("Exemplos de SPAM:")
print(data[data["label"]=="spam"]["text"].head(5).to_string(index=False))

print("\nExemplos de HAM (não spam):")
print(data[data["label"]=="ham"]["text"].head(5).to_string(index=False))
```
